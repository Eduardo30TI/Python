```python
import pandas as pd
from glob import glob
from CEP import CEP
from Acentos import Acentuacao

pd.set_option('display.max_columns',None)
```

```python
dados=glob('*.csv')

dados
```

```python
temp_df=pd.DataFrame()

for d in dados:
    
    df=pd.read_csv(d,encoding='utf-8')
    
    df['Arquivo']=d
    
    temp_df=pd.concat([temp_df,df],axis=0,ignore_index=True)
    
    pass
```

```python
len(temp_df['CNPJ'].unique().tolist())
```

```python
cep_lista=glob('cep região.xlsx')

cep_df=pd.read_excel(cep_lista[0])

cep_df['CEP Inicio']=cep_df['CEP'].apply(lambda info: info[:info.find('a')].strip())

cep_df['CEP Fim']=cep_df['CEP'].apply(lambda info: info[info.find('a'):].replace('a','').strip())

cep_df['CEP ID']=cep_df['CEP Inicio'].apply(lambda info: info[:3])

cep_df.rename(columns={'Bairro':'CEP Bairro'},inplace=True)

cep_df
```

```python
temp_df['CEP']=temp_df['CEP'].astype('str')

temp_df['CEP']=temp_df['CEP'].apply(lambda info: '0'+info if len(info)==7 else info)

temp_df['CEP Base']=temp_df['CEP'].apply(lambda info: info[:5])

temp_df['CEP ID']=temp_df['CEP Base'].apply(lambda info: info[:3])

cep_df=cep_df[['Região','CEP Bairro','CEP ID']]

temp_df=temp_df.merge(cep_df,on='CEP ID',how='left')[['Nome fantasia', 'Razão social', 'CNPJ', 'CNPJ matriz',
       'Tipo de unidade', 'CNAE principal descrição', 'Segmento',
       'Faturamento', 'Funcionários', 'Capital social', 'Porte RF',
       'Natureza jurídica', 'Natureza jurídica descrição', 'Filiais',
       'Idade empresa', 'Data de abertura', 'Nível de atividade',
       'Tendência de crescimento', 'País', 'Estado', 'Cidade', 'Bairro',
       'Endereço', 'CEP', 'Perfil instagram', 'Website', 'Possui telefone',
       'Possui e-mail', 'UF', 'Faturamento - Faixas', 'Funcionários - Faixas',
       'Porte', 'Relacionamento', 'Situação', 'Opção MEI', 'Opção Simples',
       'Faturamento grupo', 'CNAE principal', 'Funcionários grupo',
       'Telefones', 'Telefone principal', 'CNAEs secundários', 'E-mails',
       'Sócios', 'Grupo CNAE', 'Grupo CNAE código', 'Regiões',
       'Faturamento - Faixas Personalizadas', 'Palavra Mapeada',
       'Possui Instagram?', 'Tipo Logradouro', 'Endereço Número',
       'Endereço Complemento', 'CEP Base', 'CEP ID','Região','CEP Bairro','Arquivo']]
```
