```python
from glob import glob
import pandas as pd

pd.set_option('display.max_columns',None)
```

```python
arquivos=glob('*.xlsx')
```

```python
arquivos
```

```python
df=pd.read_excel(arquivos[-1])
```

```python
def Classificacao(val):
    
    if val<=33.33:
        
        tipo='A'
        
        pass
    
    elif val<=66.66:
        
        tipo='B'
        
        pass
    
    else:
        
        tipo='C'
        
        
        pass
    
    return tipo
    
    pass
```

```python
anos=[l for l in df.columns.tolist() if str(l).isnumeric()]

colunas=df.columns.tolist()

for a in anos:
    
    indixe=colunas.index(a)+1
    
    df.sort_values(a,ascending=False,ignore_index=True,inplace=True)
    
    df[f'Total {a}']=df[a].sum()
    
    total=0
    
    valores=[]
    
    for i in range(0,len(df)):
        
        total+=df.loc[i,a].sum()
        
        valores.append(total)
                
        pass
    
    df[f'Acumulado {a}']=valores
    
    df[f'Perc {a}']=round(df[f'Acumulado {a}']/df[f'Total {a}'],4)*100
    
    df[f'Classificação {a}']=df[f'Perc {a}'].apply(Classificacao)
        
    #break
    
    pass

for c in df.columns.tolist():
    
    if c in colunas or str(c).find('Classificação')>=0:
                        
        continue
        
    df.drop(columns=[c],inplace=True)
    
    pass

df
```

```python
df.to_excel('Desempenho.xlsx',encoding='UTF-8',index=False)
```
