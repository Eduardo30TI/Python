```python
for i in range(0,len(df)):
    
    ddd=df.loc[i,'DDD']
    
    telefone=df.loc[i,'Contato']
    
    email=df.loc[i,'E-mail Cliente']
    
    codigo=df.loc[i,'ID Cliente']
    
    tipo= 'CO' if len(str(telefone))==8 else 'CE'
    
    querys={
        
        
        'Cliente':
        
        """
        
        UPDATE cliente
        SET e_mail='{0}'
        WHERE cd_clien={1}
                
        """.format(email,codigo),
        
        'Telefone':
        
        """
        
        UPDATE tel_cli
        SET tp_tel='{0}',
        ddd='{1}',
        numero={2}
        WHERE cd_clien={3} AND seq=1
        
        """.format(tipo,ddd,telefone,codigo)
        
    }
    
    for query in querys.values():
        
        try:
        
            conecta=sql.ConexaoSQL()

            print(query)

            sql.Salvar(query,conecta)
            
            pass
        
        except:
            
            continue
        
        pass
    
    #break
    
    pass
```
