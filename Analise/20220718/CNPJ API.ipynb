```python
from CNPJ import CNPJ
import pandas as pd
from Acesso import Login
from ConectionSQL import SQL

pd.set_option('display.max_columns',None)

s=Login()

sql=SQL(s.usuario,s.senha,s.database,s.server)

conectando=sql.ConexaoSQL()
```

```python
querys={
    
    'Cliente':
    
    
    """
    
    SELECT * FROM netfeira.vw_cliente
    
    """
    
    
}
```

```python
tabelas_df=dict()

for tabela,query in querys.items():
    
    tabelas_df[tabela]=pd.read_sql(query,conectando)
    
    
    pass
```

```python
tabelas_df['Cliente'].columns
```

```python
cliente_df=tabelas_df['Cliente'].loc[(tabelas_df['Cliente']['CNPJ Caracter']==14)&(tabelas_df['Cliente']['Status do Cliente']=='ATIVO')]
```

```python
len(cliente_df)
```

```python
temp_df=pd.DataFrame()

for indice,linha in cliente_df.iterrows():
    
    info=CNPJ(linha['CNPJ'])
    
    dados=info.GetDados()
    
    df=pd.json_normalize(dados)
    
    temp_df=pd.concat([temp_df,df],axis=0,ignore_index=True)
            
    if(indice>=500):
        
        break
    
    pass
```

```python
temp_df.columns
```

```python
df=temp_df[['cnpj','porte','descricao_situacao_cadastral','descricao_motivo_situacao_cadastral','cnae_fiscal','cnae_fiscal_descricao']]

df=df.loc[df['descricao_situacao_cadastral'].str.contains('BAIXADA')]

df.columns=df.columns.str.title().tolist()

df['Cnae_Fiscal_Descricao']=df['Cnae_Fiscal_Descricao'].str.upper()
```

```python
cliente_df=cliente_df.merge(df,left_on='CNPJ',right_on='Cnpj',how='inner')

cliente_df.to_excel('Clientes Baixados.xlsx',index=False,encoding='UTF-8')
```
