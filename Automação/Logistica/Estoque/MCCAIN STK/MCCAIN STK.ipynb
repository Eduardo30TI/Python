```python
from ConectionSQL import SQL
from Acesso import Login
import pandas as pd

s=Login()

sql=SQL(s.usuario,s.senha,s.database,s.server)

conectando=sql.ConexaoSQL()
```

```python
querys={
    
    'Produto':
    
    """
    
    SELECT * FROM netfeira.vw_produto
    WHERE [Status]='ATIVO'    
    
    """,
    
    'Estoque':
    
    
    """
    
    SELECT * FROM netfeira.vw_estoque
    
    """
    
}
```

```python
tabelas_df=dict()

for tabela,query in querys.items():
    
    tabelas_df[tabela]=pd.read_sql(query,conectando)
    
    pass
```

```python
tabelas_df['Produto']=tabelas_df['Produto'][['SKU','Categoria','Linha','Grupo MCCAIN']].loc[tabelas_df['Produto']['Fabricante']=='MCCAIN']
```

```python
tabelas_df['Estoque']=tabelas_df['Estoque'].merge(tabelas_df['Produto'],on='SKU',how='inner')[['Local', 'Tipo', 'SKU', 'Cód. Fabricante', 'Produto', 'Fabricante','Grupo MCCAIN',
       'Unid. CMP', 'Qtde CMP']]
```

```python
estoque_df=tabelas_df['Estoque'].groupby(['SKU','Produto','Fabricante','Grupo MCCAIN','Unid. CMP'],as_index=False).sum()

grupo=estoque_df['Grupo MCCAIN'].unique().tolist()

for g in grupo:
    
    temp_df=estoque_df.loc[(estoque_df['Grupo MCCAIN']==g)&(estoque_df['Qtde CMP']>0)]
    
    display(temp_df)
    
    pass
```
