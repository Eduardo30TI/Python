```python
import tabula
import PyPDF2
import warnings
from Acesso import Login
from Query import Query
from Interface import GUI
from pathlib import Path as base_path
import os
import pandas as pd

gui=GUI()

s=Login()

sql=Query(s.usuario,s.senha,s.database,s.server)

path_foto='Foto Empresa'

querys={

    'Produto':

    """

    SELECT p.SKU,p.[Cód. Fabricante],p.Produto,p.EAN,p.Fabricante,p.Linha
    FROM netfeira.vw_produto p
    ORDER BY Produto
    
    """
}

warnings.filterwarnings('ignore')
```

```python
#carregando arquivo

gui.Cls()

paths_dict={'Origem':'','Destino':''}

for c in paths_dict.keys():

    while True:

        caminho=input(f'Informe o caminho de {str(c).lower()}: ')

        if caminho!='' and gui.PathExists(caminho)==True:

            paths_dict[c]=caminho

            break

        pass

    pass
    
dirs=base_path(paths_dict['Origem'])

lista=list(dirs.rglob('*.pdf*'))
```

```python
print(len(lista))
```

```python
df=sql.CriarTabela(kwargs=querys)
```

```python
def ConverterFloat(val: str):

    val=str(val)

    val=val.strip().split('.')

    return val[0]

    pass
```

```python
codigos=[]

dados_dict=dict()

for i,arqs in enumerate(lista):

    try:

        arq=gui.BaseName(arq)

        print(f'Lendo arquivo {i+1} de {len(lista)}. Aguarde...')

        lista_pdf=tabula.read_pdf(arqs,pages='all')

        for l in lista_pdf:

            colunas=l.columns.tolist()

            if str(colunas[0]).lower()!='código interno do produto':

                continue

            print(l)
                    
            temp_df=l

            temp_df['CÓDIGO INTERNO DO PRODUTO']=temp_df['CÓDIGO INTERNO DO PRODUTO'].apply(ConverterFloat)

            for c in temp_df.loc[temp_df['CÓDIGO INTERNO DO PRODUTO'].notnull(),'CÓDIGO INTERNO DO PRODUTO'].unique().tolist():

                if str(c).find('/')>=0:

                    ids=str(c).strip().split('/')

                    for id in ids:

                        codigos.append(id)

                        dados_dict[id]=arqs

                        pass

                    pass

                else:

                    if c in codigos:

                        continue
                        
                    codigos.append(c)

                    dados_dict[c]=arqs

                    pass

                pass

            pass

        pass

    except:

        continue

    #break

    pass
```
